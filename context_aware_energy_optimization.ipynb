# Context-Aware Smart Energy Optimization System

## Problem Statement
Energy consumption in households varies based on user behavior and time patterns. Identifying these patterns can help optimize energy usage.

## Objective
To analyze energy consumption data, detect behavioral patterns, and suggest optimized usage strategies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Data Simulation

Simulating smart home energy usage data based on daily patterns.

```python
np.random.seed(42)

time = np.arange(0, 200)

# Base energy usage
energy = np.random.normal(loc=3, scale=0.5, size=200)

# Simulate high usage (evening peak)
energy[50:80] += 2
energy[150:170] += 2.5

df = pd.DataFrame({
    'Time': time,
    'Energy_Consumption': energy
})

df.head()
```

## Visualization of Energy Consumption

```python
plt.figure(figsize=(12,5))
plt.plot(df['Time'], df['Energy_Consumption'], label='Energy Usage')
plt.title("Energy Consumption Over Time")
plt.xlabel("Time")
plt.ylabel("Energy (kWh)")
plt.legend()
plt.show()
```

## Behavioral Pattern Detection

```python
# Categorize usage behavior
def classify_usage(x):
    if x < 3:
        return "Low Usage"
    elif x < 4.5:
        return "Normal Usage"
    else:
        return "High Usage"

df['Usage_Type'] = df['Energy_Consumption'].apply(classify_usage)

df.head()
```

## Visualization of Usage Behavior

```python
colors = df['Usage_Type'].map({
    'Low Usage': 'green',
    'Normal Usage': 'orange',
    'High Usage': 'red'
})

plt.figure(figsize=(12,5))
plt.scatter(df['Time'], df['Energy_Consumption'], c=colors)
plt.title("Energy Usage Classification")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.show()
```

## Optimization Suggestions

```python
# Identify peak usage periods
peak_times = df[df['Usage_Type'] == "High Usage"]

print("High energy usage detected at these time intervals:")
print(peak_times[['Time', 'Energy_Consumption']].head())

print("\nSuggestions:")
print("- Shift heavy appliance usage to low-demand periods")
print("- Use smart scheduling for energy efficiency")
```

## Conclusion

The system successfully identifies energy consumption patterns and provides optimization suggestions. This approach can be extended to real-world smart home systems.
